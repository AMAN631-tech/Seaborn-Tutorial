# Data Visualization With Seaborn.

# What is Seaborn?
Seaborn is a Python library that is used for data visualization.

It is built on top of Matplotlib, which means it uses Matplotlib internally but makes the code simpler and more beautiful.

With Seaborn, you can create attractive charts very easily.

# Why use Seaborn?
It reduces the complexity of writing long code in Matplotlib.

You can make complex statistical plots (like regression lines, heatmaps, etc.) in just a few lines.

Seaborn has better default styles—the graphs automatically look more professional.



```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot..

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatterplot of Total Bill vs Tip.')
plt.show()
```

```python
# line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line plot for total bill based on size.')
plt.show()
```

```python
# Creating bar plot.

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Total bill by day')
plt.show()
```

# explaination.

| Part                       | Meaning                                                                                        |
| -------------------------- | ---------------------------------------------------------------------------------------------- |
| **Blue bar**               | The **average bill (mean)** of your data                                                       |
| **Black line (Error bar)** | **Not the highest or lowest bill** – It shows the **range of where the real average could be** |
| **Top of black line**      | **Upper guess of the real average** (e.g., 115)                                                |
| **Bottom of black line**   | **Lower guess of the real average** (e.g., 105)                                                |


```python
# Creating box plot.

sns.boxplot(x='day',y='total_bill',data=tips)

plt.show()
```

```python
# creating violin plot
sns.set_theme(style="darkgrid")
sns.violinplot(x='day',y='total_bill',data=tips)

plt.show()
```

```python
# histogram

sns.histplot(x='total_bill',bins=10,data=tips,kde=True)
```

```python
import numpy as np

# Get the data for total_bill
data = tips['total_bill']

# Use numpy to calculate bins
counts, bin_edges = np.histogram(data, bins=10)

# Print the bin edges
print("Bin edges:", bin_edges)

# Show ranges of each bin
for i in range(len(bin_edges)-1):
    print(f"Bin {i+1}: {bin_edges[i]} to {bin_edges[i+1]}")
```

```python
# kde plot
sns.kdeplot(x=tips['total_bill'],fill=True)
```

```python
# pairplot

sns.pairplot(tips)
```

```python
numeric_data = tips.select_dtypes(['int64','float64'])
numeric_data
```

```python
corr = numeric_data.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='YlGnBu')
```
